```python
import numpy as np

a = np.array([0.3, 2.9, 4.0])
exp_a = np.exp(a)

print(exp_a)
```

```python
sum_a = np.sum(exp_a)
dis_a = exp_a / sum_a

print(dis_a)
```

```python
print(np.sum(dis_a))
```

```python
def softmax(a):
    exp_a = np.exp(a)
    sum_exp_a = np.sum(exp_a)

    return exp_a / sum_exp_a

def show_softmax(a):
    print(softmax(a))

show_softmax(np.array([0.3, 2.9, 4.0]))

## error
show_softmax(np.array([1010, 1000, 990]))
```

```python
def softmax(a):
    c = np.max(a)
    # overflow handle
    exp_a = np.exp(a - c)

    sum_exp_a = np.sum(exp_a)
    return exp_a / sum_exp_a

y = softmax(np.array([0.3, 2.9, 4.0]))
print(y)
print(np.sum(y))
```
